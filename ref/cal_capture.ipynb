```python
import cv2
import os
import numpy as np
```

```python
SOURCE = "stereo_calibration_video.avi"
SAVE_LDIR = "./cal/left"
SAVE_RDIR = "./cal/right"

CHECKERBOARD = (9, 6)
FRAME_INTERVAL = 15
```

```python
cap = cv2.VideoCapture(SOURCE)
total_frame = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print(f"영상 로드: {total_frame} 프레임")

frame_idx = 0
saved_cnt = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    if frame_idx % FRAME_INTERVAL == 0:
        
        height, width, _ = frame.shape
        half_width = width // 2

        imgL = frame[:, :half_width]
        imgR = frame[:, half_width:]

        grayL = cv2.cvtColor(imgL, cv2.COLOR_BGR2GRAY)
        grayR = cv2.cvtColor(imgR, cv2.COLOR_BGR2GRAY)

        retL, cornersL = cv2.findChessboardCorners(grayL, CHECKERBOARD, None)
        retR, cornersR = cv2.findChessboardCorners(grayR, CHECKERBOARD, None)

        if retL and retR:
            save_name_L = os.path.join(SAVE_LDIR, f"l_{int(saved_cnt)}.jpg")
            save_name_R = os.path.join(SAVE_RDIR, f"r_{int(saved_cnt)}.jpg")

            cv2.imwrite(save_name_L, imgL)
            cv2.imwrite(save_name_R, imgR)

            saved_cnt += 1

    frame_idx += 1

cap.release()
cv2.destroyAllWindows()
```
